```python
import numpy as np
import matplotlib.pyplot as plt
import io
import math
import pandas as pd
```

```python
df_cms = pd.read_csv('cmsdata.txt', sep=' ')
mass_cms_min = df_cms["q2lo"]
mass_cms_max = df_cms["q2hi"]
mass_cms = (mass_cms_min + mass_cms_max) / 2.
ds_dm_cms = df_cms["PDF0"] # pb-1
err_ds_dm_cms = df_cms["uncertainty"] # pb-1
ds_dm_cms_min = ds_dm_cms - err_ds_dm_cms
ds_dm_cms_max= ds_dm_cms + err_ds_dm_cms

df_atlas = pd.read_csv('atlasdata.txt', sep=' ')
mass_atlas_min = df_atlas["q2lo"]
mass_atlas_max = df_atlas["q2hi"]
mass_atlas = (mass_atlas_min + mass_atlas_max) / 2.
ds_dm_atlas = df_atlas["PDF0"] # pb-1
err_ds_dm_atlas = df_atlas["uncertainty"] # pb-1
ds_dm_atlas_min = ds_dm_atlas - err_ds_dm_atlas
ds_dm_atlas_max= ds_dm_atlas + err_ds_dm_atlas
```

```python
df_powheg_13TeV = pd.read_csv('powheg_13TeV_mass.txt', sep=' ')
mass_powheg_13TeV_min = df_powheg_13TeV["q2lo"]
mass_powheg_13TeV_max = df_powheg_13TeV["q2hi"]
mass_powheg_13TeV_width = mass_powheg_13TeV_max - mass_powheg_13TeV_min
mass_powheg_13TeV = (mass_powheg_13TeV_min + mass_powheg_13TeV_max) / 2.
ds_dm_powheg_13TeV = df_powheg_13TeV["PDF0"] # pb-1
err_ds_dm_powheg_13TeV = df_powheg_13TeV["uncertainty"] # pb-1
ds_dm_powheg_13TeV_min = ds_dm_powheg_13TeV - err_ds_dm_powheg_13TeV
ds_dm_powheg_13TeV_max= ds_dm_powheg_13TeV + err_ds_dm_powheg_13TeV

df_dyturbo_13TeV = pd.read_csv('dyturbo_13TeV_mass.txt', sep=' ')
mass_dyturbo_13TeV_min = df_dyturbo_13TeV["q2lo"]
mass_dyturbo_13TeV_max = df_dyturbo_13TeV["q2hi"]
mass_dyturbo_13TeV_width = mass_dyturbo_13TeV_max - mass_dyturbo_13TeV_min
mass_dyturbo_13TeV = (mass_dyturbo_13TeV_min + mass_dyturbo_13TeV_max) / 2.
ds_dmass_dyturbo_13TeV = df_dyturbo_13TeV["PDF0"] * (1e-3 / mass_dyturbo_13TeV_width) # pb-1
err_ds_dmass_dyturbo_13TeV = df_dyturbo_13TeV["uncertainty"] * (1e-3 / mass_dyturbo_13TeV_width) # pb-1
ds_dmass_dyturbo_13TeV_min = ds_dmass_dyturbo_13TeV - err_ds_dmass_dyturbo_13TeV
ds_dmass_dyturbo_13TeV_max= ds_dmass_dyturbo_13TeV + err_ds_dmass_dyturbo_13TeV


# Creating histogram
plt.title("CMS comparison with POWHEG and DYTURBO") 
plt.xlabel("m (GeV)") 
plt.ylabel("d\u03C3/dm (pb/GeV)") 
#plt.xscale('log')
plt.yscale('log')
plt.title(r'$d\sigma_{Z / \gamma*}/dm, \sqrt{s} = 13 TeV$')
plt.errorbar(mass_cms, ds_dm_cms, yerr=0, label="CMS", fmt='o', color='black', ecolor='black', alpha=1, elinewidth=10, capsize=0)
plt.errorbar(mass_powheg_13TeV, ds_dm_powheg_13TeV, yerr=0, label="POWHEG", fmt='--', color='blue', ecolor='blue', alpha=0.9, elinewidth=10, capsize=0)
#plt.fill_between(mass_dyturbo_13TeV, ds_dmass_dyturbo_13TeV_min, ds_dmass_dyturbo_13TeV_max, label="dyturbo", alpha=0.4, edgecolor='red', facecolor='red', hatch='////', linewidth=1, linestyle='dashdot', antialiased=True)
plt.errorbar(mass_dyturbo_13TeV, ds_dmass_dyturbo_13TeV, yerr=0, label="DYTurbo", fmt='--', color='red', ecolor='red', alpha=0.9, elinewidth=10, capsize=0)
plt.legend(loc='upper center')
plt.show()
```

```python
df_powheg_7TeV = pd.read_csv('powheg_7TeV_mass.txt', sep=' ')
mass_powheg_7TeV_min = df_powheg_7TeV["q2lo"]
mass_powheg_7TeV_max = df_powheg_7TeV["q2hi"]
mass_powheg_7TeV_width = mass_powheg_7TeV_max - mass_powheg_7TeV_min
mass_powheg_7TeV = (mass_powheg_7TeV_min + mass_powheg_7TeV_max) / 2.
ds_dm_powheg_7TeV = df_powheg_7TeV["PDF0"] # pb-1
err_ds_dm_powheg_7TeV = df_powheg_7TeV["uncertainty"] # pb-1
ds_dm_powheg_7TeV_min = ds_dm_powheg_7TeV - err_ds_dm_powheg_7TeV
ds_dm_powheg_7TeV_max= ds_dm_powheg_7TeV + err_ds_dm_powheg_7TeV

df_dyturbo_7TeV = pd.read_csv('dyturbo_7TeV_mass.txt', sep=' ')
mass_dyturbo_7TeV_min = df_dyturbo_7TeV["q2lo"]
mass_dyturbo_7TeV_max = df_dyturbo_7TeV["q2hi"]
mass_dyturbo_7TeV_width = mass_dyturbo_7TeV_max - mass_dyturbo_7TeV_min
mass_dyturbo_7TeV = (mass_dyturbo_7TeV_min + mass_dyturbo_7TeV_max) / 2.
ds_dmass_dyturbo_7TeV = df_dyturbo_7TeV["PDF0"] * (1e-3 / mass_dyturbo_7TeV_width) # pb-1
err_ds_dmass_dyturbo_7TeV = df_dyturbo_7TeV["uncertainty"] * (1e-3 / mass_dyturbo_7TeV_width) # pb-1
ds_dmass_dyturbo_7TeV_min = ds_dmass_dyturbo_7TeV - err_ds_dmass_dyturbo_7TeV
ds_dmass_dyturbo_7TeV_max= ds_dmass_dyturbo_7TeV + err_ds_dmass_dyturbo_7TeV

# Creating histogram
plt.title("ATLAS comparison with POWHEG & DYTURBO") 
plt.xlabel("m (GeV)") 
plt.ylabel("d\u03C3/dm (pb/GeV)") 
#plt.xscale('log')
plt.yscale('log')
plt.title(r'$d\sigma_{Z / \gamma*}/dm, \sqrt{s} = 7 TeV$')
plt.errorbar(mass_atlas, ds_dm_atlas, yerr=0, label="ATLAS", fmt='o', color='black', ecolor='black', alpha=1, elinewidth=10, capsize=0)
plt.errorbar(mass_powheg_7TeV, ds_dm_powheg_7TeV, yerr=0, label="POWHEG", fmt='--', color='blue', ecolor='blue', alpha=0.9, elinewidth=10, capsize=0)
plt.errorbar(mass_dyturbo_7TeV, ds_dmass_dyturbo_7TeV, yerr=0, label="DYTurbo", fmt='--', color='red', ecolor='red', alpha=0.9, elinewidth=10, capsize=0)
plt.legend(loc='upper center')
plt.show()
```

```python
df_dyturbo_7TeV_vs_y = pd.read_csv('dyturbo_7TeV_y.txt', sep=' ')
y_dyturbo_7TeV_min = df_dyturbo_7TeV_vs_y["ylo"]
y_dyturbo_7TeV_max = df_dyturbo_7TeV_vs_y["yhi"]
y_dyturbo_7TeV = (y_dyturbo_7TeV_min + y_dyturbo_7TeV_max) / 2.
ds_dy_dyturbo_7TeV = df_dyturbo_7TeV_vs_y["PDF0"] * (1e-3 ) # pb-1
err_ds_dy_dyturbo_7TeV = df_dyturbo_7TeV_vs_y["uncertainty"] * (1e-3) # pb-1
ds_dy_dyturbo_7TeV_min = ds_dy_dyturbo_7TeV - err_ds_dy_dyturbo_7TeV
ds_dy_dyturbo_7TeV_max= ds_dy_dyturbo_7TeV + err_ds_dy_dyturbo_7TeV

df_dyturbo_13TeV_vs_y = pd.read_csv('dyturbo_13TeV_y.txt', sep=' ')
y_dyturbo_13TeV_min = df_dyturbo_13TeV_vs_y["ylo"]
y_dyturbo_13TeV_max = df_dyturbo_13TeV_vs_y["yhi"]
y_dyturbo_13TeV = (y_dyturbo_13TeV_min + y_dyturbo_13TeV_max) / 2.
ds_dy_dyturbo_13TeV = df_dyturbo_13TeV_vs_y["PDF0"] * (1e-3 ) # pb-1
err_ds_dy_dyturbo_13TeV = df_dyturbo_13TeV_vs_y["uncertainty"] * (1e-3) # pb-1
ds_dy_dyturbo_13TeV_min = ds_dy_dyturbo_13TeV - err_ds_dy_dyturbo_13TeV
ds_dy_dyturbo_13TeV_max= ds_dy_dyturbo_13TeV + err_ds_dy_dyturbo_13TeV

plt.title("DYTURBO vs y") 
plt.xlabel("m (GeV)") 
plt.ylabel("d\u03C3/dm (pb/GeV)") 
plt.title(r'$d\sigma_{Z / \gamma*}/dm, DYTurbo$')
plt.errorbar(y_dyturbo_7TeV, ds_dy_dyturbo_7TeV, yerr=err_ds_dy_dyturbo_7TeV, label="DYTurbo 7 TeV", fmt='--', color='red', ecolor='red', alpha=0.9, elinewidth=10, capsize=0)
plt.errorbar(y_dyturbo_13TeV, ds_dy_dyturbo_13TeV, yerr=err_ds_dy_dyturbo_13TeV, label="DYTurbo 13 TeV", fmt='--', color='blue', ecolor='blue', alpha=0.9, elinewidth=10, capsize=0)
plt.legend(loc='upper right')
plt.show()
```
