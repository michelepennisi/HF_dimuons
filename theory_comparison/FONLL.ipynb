```python
# Si semplificano delle incertezze in FONLL & NLO??
ratio_cc_bb_FONLL_NNPDF = ds_dy_cc_FONLL_NNPDF / ds_dy_bb_FONLL_NNPDF
err_min_ratio_cc_bb_FONLL_NNPDF = ratio_cc_bb_FONLL_NNPDF * ((ds_dy_cc_FONLL_NNPDF - min_ds_dy_cc_FONLL_NNPDF) / ds_dy_cc_FONLL_NNPDF)
err_max_ratio_cc_bb_FONLL_NNPDF = ratio_cc_bb_FONLL_NNPDF * ((max_ds_dy_cc_FONLL_NNPDF - ds_dy_cc_FONLL_NNPDF) / ds_dy_cc_FONLL_NNPDF)

ratio_cc_bb_FONLL_CTEQ6 = ds_dy_cc_FONLL_CTEQ6 / ds_dy_bb_FONLL_CTEQ6
err_min_ratio_cc_bb_FONLL_CTEQ6 = ratio_cc_bb_FONLL_CTEQ6 * ((ds_dy_cc_FONLL_CTEQ6 - min_ds_dy_cc_FONLL_CTEQ6) / ds_dy_cc_FONLL_CTEQ6)
err_max_ratio_cc_bb_FONLL_CTEQ6 = ratio_cc_bb_FONLL_CTEQ6 * ((max_ds_dy_cc_FONLL_CTEQ6 - ds_dy_cc_FONLL_CTEQ6) / ds_dy_cc_FONLL_CTEQ6)

ratio_cc_bb_NLO_NNPDF = ds_dy_cc_NLO_NNPDF / ds_dy_bb_NLO_NNPDF
err_min_ratio_cc_bb_NLO_NNPDF = ratio_cc_bb_NLO_NNPDF * ((ds_dy_cc_NLO_NNPDF - min_ds_dy_cc_NLO_NNPDF) / ds_dy_cc_NLO_NNPDF)
err_max_ratio_cc_bb_NLO_NNPDF = ratio_cc_bb_NLO_NNPDF * ((max_ds_dy_cc_NLO_NNPDF - ds_dy_cc_NLO_NNPDF) / ds_dy_cc_NLO_NNPDF)

ratio_cc_bb_NLO_CTEQ6 = ds_dy_cc_NLO_CTEQ6 / ds_dy_bb_NLO_CTEQ6
err_min_ratio_cc_bb_NLO_CTEQ6 = ratio_cc_bb_NLO_CTEQ6 * ((ds_dy_cc_NLO_CTEQ6 - min_ds_dy_cc_NLO_CTEQ6) / ds_dy_cc_NLO_CTEQ6)
err_max_ratio_cc_bb_NLO_CTEQ6 = ratio_cc_bb_NLO_CTEQ6 * ((max_ds_dy_cc_NLO_CTEQ6 - ds_dy_cc_NLO_CTEQ6) / ds_dy_cc_NLO_CTEQ6)

ratio_cc_bb_PYTHIA = ds_cc_dy_PYTHIA / ds_bb_dy_PYTHIA
sq_rel_stat_ds_cc_dy_PYTHIA = (stat_ds_cc_dy_PYTHIA / ds_cc_dy_PYTHIA) * (stat_ds_cc_dy_PYTHIA / ds_cc_dy_PYTHIA)
sq_rel_stat_ds_bb_dy_PYTHIA = (stat_ds_bb_dy_PYTHIA / ds_bb_dy_PYTHIA) * (stat_ds_bb_dy_PYTHIA / ds_bb_dy_PYTHIA)
sq_rel_syst_ds_cc_dy_PYTHIA = (syst_ds_cc_dy_PYTHIA / ds_cc_dy_PYTHIA) * (syst_ds_cc_dy_PYTHIA / ds_cc_dy_PYTHIA)
sq_rel_syst_ds_bb_dy_PYTHIA = (syst_ds_bb_dy_PYTHIA / ds_bb_dy_PYTHIA) * (syst_ds_cc_dy_PYTHIA / ds_cc_dy_PYTHIA)
err_ratio_cc_bb_PYTHIA = ratio_cc_bb_PYTHIA * np.sqrt(sq_rel_stat_ds_cc_dy_PYTHIA + sq_rel_stat_ds_bb_dy_PYTHIA + sq_rel_syst_ds_cc_dy_PYTHIA + sq_rel_syst_ds_bb_dy_PYTHIA)

plt.title("FONLL & NLO: charm / beauty") 
plt.xlabel("y") 
plt.ylabel("ratio") 
plt.yscale('log')
plt.title(r'($d\sigma_{c\bar{c}}/dy$) / ($d\sigma_{b\bar{b}}/dy$)') 
plt.errorbar(y_FONLL_NNPDF, ratio_cc_bb_FONLL_NNPDF, yerr=0, label="FONLL NNPDF30", fmt='--', color='red', ecolor='red', alpha=0.3, elinewidth=10, capsize=0)
plt.errorbar(y_NLO_NNPDF, ratio_cc_bb_NLO_NNPDF, yerr=0, label="NLO NNPDF30", fmt='--', color='blue', ecolor='blue', alpha=0.3, elinewidth=10, capsize=0)
#plt.fill_between(y_FONLL_NNPDF, err_min_ratio_cc_bb_FONLL_NNPDF, err_max_ratio_cc_bb_FONLL_NNPDF, label="FONLL NNPDF30 unc. tot", alpha=0.4, edgecolor='gray', facecolor='gray', hatch='////', linewidth=1, linestyle='dashdot', antialiased=True)
plt.errorbar(y_FONLL_CTEQ6, ratio_cc_bb_FONLL_CTEQ6, yerr=0, label="FONLL CTEQ6.6", fmt='-', color='red', ecolor='red', alpha=0.3, elinewidth=10, capsize=0)
plt.errorbar(y_NLO_CTEQ6, ratio_cc_bb_NLO_CTEQ6, yerr=0, label="NLO CTEQ6.6", fmt='-', color='blue', ecolor='blue', alpha=0.3, elinewidth=10, capsize=0)
#plt.fill_between(y_FONLL_CTEQ6, err_min_ratio_cc_bb_FONLL_CTEQ6, err_max_ratio_cc_bb_FONLL_CTEQ6, label="FONLL CTEQ6.6 unc. tot", alpha=0.4, edgecolor='gray', facecolor='gray', linewidth=1, linestyle='dashdot', antialiased=True)
plt.errorbar(y, ratio_cc_bb_PYTHIA, xerr=dy, yerr=err_ratio_cc_bb_PYTHIA, label="ALICE", fmt='o', color='black', ecolor='black')
plt.legend(loc='upper center')
plt.show()
```
