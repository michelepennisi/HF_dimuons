```python
import numpy as np
import matplotlib.pyplot as plt
import io
import math
import pandas as pd
```

```python
df_FONLL_cc_NNPDF = pd.read_csv('FONLL_cc_Pt_0_30_NNPDF.txt', sep=' ')
y_FONLL_NNPDF = df_FONLL_cc_NNPDF["y"]
ds_dy_cc_FONLL_NNPDF = df_FONLL_cc_NNPDF["central"] * 1e-06
min_ds_dy_cc_FONLL_NNPDF = df_FONLL_cc_NNPDF["min"] * 1e-06
max_ds_dy_cc_FONLL_NNPDF = df_FONLL_cc_NNPDF["max"] * 1e-06
min_sc_ds_dy_cc_FONLL_NNPDF = df_FONLL_cc_NNPDF["min_sc"] * 1e-06
max_sc_ds_dy_cc_FONLL_NNPDF = df_FONLL_cc_NNPDF["max_sc"] * 1e-06
min_mass_ds_dy_cc_FONLL_NNPDF = df_FONLL_cc_NNPDF["min_mass"] * 1e-06
max_mass_ds_dy_cc_FONLL_NNPDF = df_FONLL_cc_NNPDF["max_mass"] * 1e-06
min_pdf_ds_dy_cc_FONLL_NNPDF = df_FONLL_cc_NNPDF["min_pdf"] * 1e-06
max_pdf_ds_dy_cc_FONLL_NNPDF = df_FONLL_cc_NNPDF["max_pdf"] * 1e-06

df_FONLL_cc_CTQ6 = pd.read_csv('FONLL_cc_Pt_0_30_CTQ6.txt', sep=' ')
y_FONLL_CTQ6 = df_FONLL_cc_CTQ6["y"]
ds_dy_cc_FONLL_CTQ6 = df_FONLL_cc_CTQ6["central"] * 1e-06
min_ds_dy_cc_FONLL_CTQ6 = df_FONLL_cc_CTQ6["min"] * 1e-06
max_ds_dy_cc_FONLL_CTQ6 = df_FONLL_cc_CTQ6["max"] * 1e-06
min_sc_ds_dy_cc_FONLL_CTQ6 = df_FONLL_cc_CTQ6["min_sc"] * 1e-06
max_sc_ds_dy_cc_FONLL_CTQ6 = df_FONLL_cc_CTQ6["max_sc"] * 1e-06
min_mass_ds_dy_cc_FONLL_CTQ6 = df_FONLL_cc_CTQ6["min_mass"] * 1e-06
max_mass_ds_dy_cc_FONLL_CTQ6 = df_FONLL_cc_CTQ6["max_mass"] * 1e-06
min_pdf_ds_dy_cc_FONLL_CTQ6 = df_FONLL_cc_CTQ6["min_pdf"] * 1e-06
max_pdf_ds_dy_cc_FONLL_CTQ6 = df_FONLL_cc_CTQ6["max_pdf"] * 1e-06


y = np.array([0, 3.25])
dy = np.array([0.5, 0.75])

ds_cc_dy_PYTHIA = np.array([974, 1549.3])
stat_ds_cc_dy_PYTHIA = np.array([138, 20.8])
syst_ds_cc_dy_PYTHIA = np.array([140, 173.5])

# Creating histogram
plt.title("FONLL comparison NNPDF") 
plt.xlabel("y") 
plt.ylabel("d\u03C3/dy (\u03BCb)") 
plt.yscale('log')
plt.title(r'$d\sigma_{c\bar{c}}/dy$ NNPDF30') 
plt.fill_between(y_FONLL_NNPDF, min_ds_dy_cc_FONLL_NNPDF, max_ds_dy_cc_FONLL_NNPDF, label="FONLL unc. tot", alpha=0.4, edgecolor='gray', facecolor='gray', linewidth=1, linestyle='dashdot', antialiased=True)
plt.fill_between(y_FONLL_NNPDF, min_sc_ds_dy_cc_FONLL_NNPDF, max_sc_ds_dy_cc_FONLL_NNPDF, label="FONLL unc. scale", alpha=0.3, edgecolor='gray', facecolor='gray', hatch='////', linewidth=1, linestyle='dashdot', antialiased=True)
plt.fill_between(y_FONLL_NNPDF, min_mass_ds_dy_cc_FONLL_NNPDF, max_mass_ds_dy_cc_FONLL_NNPDF, label="FONLL unc. mass", alpha=0.1, edgecolor='green', facecolor='green', linewidth=1, linestyle='dashdot', antialiased=True)
plt.fill_between(y_FONLL_NNPDF, min_pdf_ds_dy_cc_FONLL_NNPDF, max_pdf_ds_dy_cc_FONLL_NNPDF, label="FONLL unc.pdf", alpha=0.1, edgecolor='blue', facecolor='blue', linewidth=1, linestyle='dashdot', antialiased=True)
plt.errorbar(y, ds_cc_dy_PYTHIA, yerr=syst_ds_cc_dy_PYTHIA, label="ALICE syst. error", fmt='o', color='black', ecolor='red', alpha=0.3, elinewidth=10, capsize=0)
plt.errorbar(y, ds_cc_dy_PYTHIA, xerr=dy, yerr=stat_ds_cc_dy_PYTHIA, label="ALICE stat. error", fmt='o', color='black', ecolor='black')
plt.legend(loc='lower center')
plt.show()

plt.title("FONLL comparison CTQ6") 
plt.xlabel("y") 
plt.ylabel("d\u03C3/dy (\u03BCb)") 
plt.yscale('log')
plt.title(r'$d\sigma_{c\bar{c}}/dy$ CTQ6.6') 
plt.fill_between(y_FONLL_CTQ6, min_ds_dy_cc_FONLL_CTQ6, max_ds_dy_cc_FONLL_CTQ6, label="FONLL unc. tot", alpha=0.4, edgecolor='gray', facecolor='gray', linewidth=1, linestyle='dashdot', antialiased=True)
plt.fill_between(y_FONLL_CTQ6, min_sc_ds_dy_cc_FONLL_CTQ6, max_sc_ds_dy_cc_FONLL_CTQ6, label="FONLL unc. scale", alpha=0.3, edgecolor='gray', facecolor='gray', hatch='////', linewidth=1, linestyle='dashdot', antialiased=True)
plt.fill_between(y_FONLL_CTQ6, min_mass_ds_dy_cc_FONLL_CTQ6, max_mass_ds_dy_cc_FONLL_CTQ6, label="FONLL unc. mass", alpha=0.1, edgecolor='green', facecolor='green', linewidth=1, linestyle='dashdot', antialiased=True)
plt.fill_between(y_FONLL_CTQ6, min_pdf_ds_dy_cc_FONLL_CTQ6, max_pdf_ds_dy_cc_FONLL_CTQ6, label="FONLL unc.pdf", alpha=0.1, edgecolor='blue', facecolor='blue', linewidth=1, linestyle='dashdot', antialiased=True)
plt.errorbar(y, ds_cc_dy_PYTHIA, yerr=syst_ds_cc_dy_PYTHIA, label="ALICE syst. error", fmt='o', color='black', ecolor='red', alpha=0.3, elinewidth=10, capsize=0)
plt.errorbar(y, ds_cc_dy_PYTHIA, xerr=dy, yerr=stat_ds_cc_dy_PYTHIA, label="ALICE stat. error", fmt='o', color='black', ecolor='black')
plt.legend(loc='lower center')
plt.show()
```
